**Fonte:** https://www.anac.gov.br/assuntos/setor-regulado/empresas/envio-de-informacoes/microdados

**Período:** dezembro de 2020

**Contexto:** A base de dados refere-se a _Etapa Combinada (On flight origin and destination - OFOD)_

* As _etapas combinadas_ identificam os pares de origem, onde ocorreu o embarque do objeto de transporte, e destino, onde ocorreu o desembarque do objeto de transporte, independentemente da existência de aeródromos intermediários, atendidos por determinado voo. É a etapa de voo vista com foco no objeto de transporte (pessoas e/ou cargas), com base no embarque e no desembarque nos aeródromos relacionados. Os dados estatísticos da etapa combinada informam a origem e destino dos passageiros e cargas transportadas no voo, independentemente das escalas realizadas.

## 5. Distribuições
[Retorna à Tabela de conteúdo](#Tabela-de-conteúdo)

```python
# boxplot
plt.figure(figsize=(16, 6))
df_clean[['nr_passag_pagos', 'nr_passag_gratis']].boxplot(patch_artist=True, boxprops=dict(facecolor='orange', color='orange'), 
               medianprops=dict(color='red'))
plt.title("Pagantes x Não pagantes", fontsize=20)
plt.show()
```

* A quantidade de passageiros pagantes apresenta maior variabilidade/dispersão.

* Entre os pagantes, as viagens concentram-se entre 0 e 85 passageiros.

```python
f, axes = plt.subplots(1, 2, figsize=(10, 6))
#sns.despine(left=True)

sns.distplot(df_clean['nr_passag_pagos'], ax=axes[0], color="orange")
sns.distplot(df_clean['nr_passag_gratis'], ax=axes[1], color="orange")

#plt.setp(axes, yticks=[])
#plt.setp(axes, xticks=[])
plt.tight_layout()
```

* O transporte de poucos passageiros concentra-se na maior parte das viagens. 

```python
# boxplot
plt.figure(figsize=(16, 10))
df_clean[['kg_bagagem_livre', 'kg_bagagem_excesso', 'kg_carga_paga', 'kg_carga_gratis', 'kg_correio']].boxplot(patch_artist=True, 
            boxprops=dict(facecolor='orange', color='orange'), medianprops=dict(color='red'))
plt.title("Bagagem/Carga x Peso(kg)", fontsize=20)
plt.show()
```

* O kg das cargas pagas apresentam a maior variabilidade - fonte de receita direta ou indireta para a empresa aérea.

```python
f, axes = plt.subplots(3, 2, figsize=(12, 8))
#sns.despine(left=True)

sns.distplot(df_clean['kg_bagagem_livre'], ax=axes[0, 0])
sns.distplot(df_clean['kg_bagagem_excesso'], ax=axes[0, 1], kde_kws={'bw': 0.5})
sns.distplot(df_clean['kg_carga_paga'], ax=axes[1, 0])
sns.distplot(df_clean['kg_carga_gratis'], ax=axes[1, 1], kde_kws={'bw': 0.5})
sns.distplot(df_clean['kg_correio'], ax=axes[2, 0], kde_kws={'bw': 0.5})

f.delaxes(ax=axes[2, 1])

#plt.setp(axes, yticks=[])
#plt.setp(axes, xticks=[])
plt.tight_layout()
```

* Predominância de bagagens dentro do limite.

💡 Carga paga gera fonte mas possui pouco volume.
